```python
# 📌 Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 📌 Upload file (IMPORTANT for Colab)
from google.colab import files
uploaded = files.upload()

# 📌 Load dataset
df = pd.read_csv("students.csv")

# 📌 Show dataset
print("🔹 First 5 Rows of Data:")
print(df.head())

# 📌 Data Cleaning
print("\n🔹 Missing Values:")
print(df.isnull().sum())

df.drop_duplicates(inplace=True)

# 📌 Create Average column
df['Average'] = (df['Math'] + df['Science'] + df['English']) / 3

# 📌 Summary Statistics
print("\n🔹 Summary Statistics:")
print(df.describe())

# 📊 Graph 1: Average Marks
print("\n📊 Graph 1: Average Marks of Students")
plt.figure()
plt.plot(df['Name'], df['Average'])
plt.title("Average Marks of Students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()

# 📊 Graph 2: Study Hours vs Marks
print("\n📊 Graph 2: Study Hours vs Average Marks")
plt.figure()
sns.barplot(x='Hours_Studied', y='Average', data=df)
plt.title("Study Hours vs Average Marks")
plt.show()

# 📊 Graph 3: Attendance vs Marks
print("\n📊 Graph 3: Attendance vs Average Marks")
plt.figure()
sns.scatterplot(x='Attendance', y='Average', data=df)
plt.title("Attendance vs Average Marks")
plt.show()

# 📌 Final Insight
print("\n🔍 INSIGHT:")
print("Students who study more hours and have higher attendance tend to score better marks.")
```
